# Data Set <a class="anchor" id="ds"></a>

**The Data set chosen from seaborn package's repository for this project is the mpg dataset.**

```python
import seaborn as sb
import matplotlib.pyplot as plt

# loading the dataset from the seaborn package.
cars=sb.load_dataset('mpg')

cars.shape #dimension of the dataset.
```

# Data Visualization <a class="anchor" id="dv"></a>

```python
cars.head() #display the first 5 rows of the dataset.
```

<u><b>Summary Table of the given mpg dataset</b></u>

```python
# Summary table as EDA
cars.describe(include='all')
```

<b><u>Displaying a overall report on the dataset using pandas_profiling.</u></b>

```python
# use minimal=True for large data set
import pandas_profiling as pdp
report3 = pdp.ProfileReport(cars, title='Pandas Profiling Report', 
                            minimal=True) #change to False for full report
report3.to_file("report3.html")
report3
```

<u><b>Plotting bar graph of nullity by column.</b></u>

```python
# missing data pack
import missingno as msno

# this will use the env of matplotlib
%matplotlib inline
msno.bar(cars)
```

<u><b>Correlogram is plotted to display the correlation between each feature.</b></u>

```python
# Correlogram - This shows the correlation between each feature
sb.pairplot(cars)
plt.show()
```

<b>Comment:</b> From the above pairplots we can decide which features are well correlated with the 'mileage per gallon' feature

<u><b>Plotting a histogram to indentify the distribution of mileage in the dataset.</b></u>

```python
# Distribution of mileage in the dataset
sb.displot(data=cars, 
            x="mpg",
            kde=True, color = 'green') 
```

<b><u>Joint plot is plotted between mpg and acceleration.</u></b>

```python
sb.jointplot(data=cars, x="mpg", y="acceleration",kind='hex')
```

<u><b>Plotting a linear regression equation between mpg and acceleration.</b></u>

```python
# Correlation between mileage per gallon  and acceleration
sb.lmplot(data=cars, 
           x="mpg", 
           y="acceleration", line_kws={'color': 'red'}
           ) #add col="time", hue="smoker"
```

<b>Comment :</b>From the above scatter plot we see a positive correlation between mileage per gallon and acceleration

<b><u>Density plot between weight and mpg.</u></b>

```python
with sb.axes_style('white'):
    sb.jointplot("weight", "mpg", data=cars, kind='kde');
```

<u><b>Plotting a linear regression equation between mpg and weight.</b></u>

```python
# Correlation between mileage per gallon  and acceleration
sb.lmplot(data=cars, 
           x="mpg", 
           y="weight", line_kws={'color': 'red'}
           ) #add col="time", hue="smoker"
```

<b>Comment :</b>From the above scatter plot we observe that with decrease in weight there is a increase in mileage per gallon

7. Now we display the scatter plot and then plot the linear regression equation between acceleration and mpg and weight and mpg. We conclude that if a user is looking for a faster car then user has to ensure a high mileage per gallon at the same time it should be a lighter body weight vehicle.
